## Find redundant indicators

```python
# find all redundant indicators
df_test = summary["index"]
indi = []
for i, name in enumerate(df_test):
    try:
        if (df_test[i][:-3]==df_test[i+1][:-7] and "ind" in df_test[i]) or (df_test[i][:-7]==df_test[i+1][:-3] and "_ct" in df_test[i]):
            indi.append(df_test[i])
    except:
        continue
```

```python
len(indi)
```

```python
redundant_object = ['cnty_cd', 'person_id_syn', 'zip_cd']
```

```python
summary["redundant_indicator"] = summary["index"].apply(lambda x: 1 if x in indi or x in redundant_object else 0)
```

```python
final = pd.read_excel("final.xlsx")
```

```python
def Find_Acronyms(x):
    for i in Acr["Acronym / Term"][1:]:
        if i.lower() == x[:len(i)]:
            return i
        
        elif x[:3]=="cci":
            return "Charlson_Comorbidity_Index"
        
        elif x[:4]=="cons":
            return "KBM_Census"
        
        elif x[:6]=="credit":
            return "Credit"
        
        elif x=="est_age":
            return "Age"
        
        elif x[:4]=="hlth":
            return "Health_Program"
        
        elif x=="hcc_weighted_sum":
            return "HCC_weighted_sum"
        
        elif x[:3]=="lab":
            return "Abnormal_lab_results"
        
        elif x=="lang_spoken_cd":
            return "Language"
        
        elif x=="mabh_seg":
            return "Medicare_segmentation"
        
        elif x[:3]=="phy":
            return "Physican"
        
        elif x[:4]=="prov":
            return "Prov"
        
        elif x[:3]=="rev":
            return "Revenue_code"
        
        elif x[:3]=="src":
            return "Humana_member"
        
        elif x[:4]=="rucc":
            return "Member_geographic_information"
        
        elif x[:6]=="smoker":
            return "Smoker"
        
        elif x[:5]=="total":
            return "Combined_Med_&_BH_values"
        
        elif x[:3]=="med":
            return "Medical_treatment"
        
        else:
            continue

final["Acronyms"] = final["index"].apply(Find_Acronyms)
```
